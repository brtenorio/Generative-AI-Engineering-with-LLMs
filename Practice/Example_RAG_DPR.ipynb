```python
import wget
import torch

import numpy as np
import random
from tqdm.auto import tqdm
from transformers import DPRQuestionEncoder, DPRQuestionEncoderTokenizer
from transformers import DPRContextEncoder, DPRContextEncoderTokenizer
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
```

```python
from typing import List
from numpy.typing import ArrayLike
```

```python
filename = 'companyPolicies.txt'
url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/6JDbUb_L3egv_eOkouY71A.txt'

# Use wget to download the file
wget.download(url, out=filename)
print('file downloaded')
```

```python
# Question Encoder
model_ckp_q = "facebook/dpr-question_encoder-single-nq-base"
question_tokenizer = DPRQuestionEncoderTokenizer.from_pretrained(model_ckp_q)
question_encoder = DPRQuestionEncoder.from_pretrained(model_ckp_q)

# Context encoder
model_ckp_ctx= "facebook/dpr-ctx_encoder-single-nq-base"
context_tokenizer = DPRContextEncoderTokenizer.from_pretrained(model_ckp_ctx)
context_encoder = DPRContextEncoder.from_pretrained(model_ckp_ctx)
```

```python
# Here I state the question I want to aks to my virtual assistent
question = "Explain me the recruitment policy at the company"

question_inputs = question_tokenizer(question, return_tensors="pt", padding=True, truncation=True, max_length=256)
question_embedding = question_encoder(**question_inputs).pooler_output.detach().numpy()

question_embedding.shape
```

```python
def open_file_context(path) -> List :
  with open(path,"r") as f:
    context_from_file = f.read().splitlines()
  return context_from_file


# define aux generate function
def aux_batch(dataset, batch_size = 64):
  for i in range(1, len(dataset), batch_size):
    batch = dataset[i:i + batch_size]
    yield batch


def encoded_context(context_from_file : List, batch_size=16) -> ArrayLike:
  list_context_encoded = []

  batch = list(aux_batch(context_from_file, batch_size=batch_size))

  progress_bar = tqdm(range(len(batch)))

  for batched_inp in batch:

    for paragraph in batched_inp:
      ctx_return_tokens = context_tokenizer(paragraph, return_tensors="pt",padding=True, truncation=True, max_length=1024)
      ctx_encoder = context_encoder(**ctx_return_tokens)
      list_context_encoded.append(ctx_encoder.pooler_output)

    progress_bar.update(1)

    # output: concatenate the list of tensors, and transform it to np.array with detach().numpy()
  return torch.cat(list_context_encoded).detach().numpy()
```

```python
# Faiss

import faiss

embedding_dim = 768

context_embedding_np = encoded_context(open_file_context('companyPolicies.txt'))

index = faiss.IndexFlatL2(embedding_dim) #L2 Euclidean distance
index.add(context_embedding_np) # makes the embedding searchable
```

```python
D,I = index.search(question_embedding, k=5) # distance, index
```

```python
print(D)
print(I)

for i_, idx in enumerate(I[0]):
  ctx_file = open_file_context('companyPolicies.txt')
  print(f"\nContext number {i_,idx}:\n {ctx_file[idx]}")
```

```python
#AutoTokenizer, AutoModelForCausalLM, pipeline

# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM

model_ckp = "openai-community/gpt2"

tokenizer = AutoTokenizer.from_pretrained(model_ckp)
model = AutoModelForCausalLM.from_pretrained(model_ckp)

# GPT2 was trained without padding. It doesn't matter as long as you give the attention mask sot it will be ignored
tokenizer.pad_token_id = 100

model.generation_config.pad_token_id = tokenizer.pad_token_id
```

```python
pipe = pipeline("text-generation", model=model, tokenizer=tokenizer)
```

```python
ctx_file = open_file_context('companyPolicies.txt')
context_extracted = [ctx_file[idx] for idx in I[0]]
input_to_model = question +"\n" +"".join(context_extracted)

print(input_to_model)
```

```python
inputs = tokenizer(input_to_model,return_tensors="pt", truncation=True, max_length=512)
inputs.input_ids[0]
tokenizer.decode(inputs.input_ids[0])
```

```python
out = model.generate(input_ids=inputs.input_ids, attention_mask = inputs.attention_mask,
                     num_beams = 4, early_stopping=True, pad_token_id=tokenizer.eos_token_id, length_penalty = 2.0 )
```
